```python
import pandas as pd
```

```python
data = pd.read_csv("data/breast_cancer_data.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data=data.drop("id",axis=1)
data=data.drop("Unnamed: 32",axis=1)
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
data.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numcols = data.dtypes[data.dtypes=='float64'].index
```

```python
for x in numcols:
    sns.boxplot(data[x],orient='h')
    plt.title(x)
    plt.show()
```

```python
for x in numcols:
    print(x)
    descr = data[x].describe()
    IQR = descr['75%'] - descr['25%']
    upper_limit =  descr['75%'] + 1.5 * IQR
    lower_limit =  descr['25%'] - 1.5 * IQR  
    #outlier replacement
    data[x][data[x]>upper_limit] = upper_limit
    data[x][data[x]<lower_limit] = lower_limit
```

```python
for x in numcols:
    sns.boxplot(data[x],orient='h')
    plt.title(x)
    plt.show()
```

```python
y = data['diagnosis']
X = data.drop("diagnosis",axis=1)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33,random_state=50)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
classifier.fit(X_train,y_train)
```

```python
ypred=classifier.predict(X_test)
```

```python
ypred
```

```python
classifier.score(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,ypred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,ypred)
```
